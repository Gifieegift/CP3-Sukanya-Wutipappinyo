```python
menu_dict = {'สุกี้':20,'ข้าวผัด':50,'ข้าวมันไก่':60}
menu_list= []
print('====Welcome====')
print(f'Our menu')
for menu , price in menu_dict.items():
        print(menu , price)
        
while True:    
    menu_input = input('please input menu_name or exit: ')
    if menu_input.lower()=="exit":
        break 
    else:
        menu_list.append([menu_input,menu_dict[menu_input]])

def showbill():
    total = 0 
    print('====bill====')
    print('====Order====')
    for i in range(len(menu_list)):
        print(menu_list[i][0], menu_list[i][1])
        total = total + int(menu_list[i][1]) 
    return total


total_bill = showbill()
print (f'total_bill: {total_bill} THB')
    
        
```
