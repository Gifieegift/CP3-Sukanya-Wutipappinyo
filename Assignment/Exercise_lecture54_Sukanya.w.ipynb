```python
username = "abc"
password = "123"

def login():
    username_input = input("username : ")
    password_input = input("password : ")
    if username_input == username and password_input == password:
        return True
    else:
        return False()
    
def showmenu():
    print("====Gift shop====")
    print("1.vat calculator")
    print("2.price calculator")
    
def menu_select():
    user_selected = int(input("Please select menu : "))
    return user_selected

def vat_cal(price,vat):
    vat = 7
    result = price + (price*vat/100)
    return result

def price_cal():
    price1 = int(input("Please input price : "))
    price2 = int(input("Please input price : "))
    total1 = vat_cal(price1,7)
    total2 = vat_cal(price2,7)
    result = total1+total2
    return result   


login()
showmenu()
menu_select()
price_cal()
```
