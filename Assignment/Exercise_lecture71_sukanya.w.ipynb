```python
menu_list = []
price_list = []

while True:
    menu_name = input('please enter menu or type exit: ')
    if menu_name.lower() == 'exit':
        break
    else:
        menu_price = int(input('plese enter price:'))
        menu_list.append(menu_name)
        price_list.append(menu_price)
print(f'menulist : {menu_list}')
print(f'price list : {price_list}')

def showbill():
    print('====bill====')
    total = 0 
    for i in range(len(menu_list)):
        print(f'{menu_list[i]} : {price_list[i]} THB')
        total = total + int(price_list[i])
    return total
total_bill = showbill()
print(f'total : {total_bill} THB')
```
