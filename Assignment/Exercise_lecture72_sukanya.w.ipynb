```python

menu_list = []
while True:
    menu_name = input('please enter menu or type exit: ')
    if menu_name.lower() == 'exit':
        break
    else:
        menu_price = int(input('plese enter price:')) 
        menu_list.append([menu_name,menu_price])
print(f'menulist : {menu_list}')

def showbill() : 
    total_price = 0 
    print('====Bill====')
    for i in range(len(menu_list)):
        print(menu_list[i][0], menu_list[i][1])
        total_price = total_price + int(menu_list[i][1])
    return total_price
 
        
sum_price = showbill()
print(f'total_price : {sum_price} THB')
```
