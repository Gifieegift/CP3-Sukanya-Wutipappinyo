```python
class Vehicle :
    license_code = ""
    serialcode = "" 
    def turn_on_air(self):
        print("Car is turning on air conditioner")
    def turn_off_air(self):
        print("Car is turning off air conditioner")

class Car(Vehicle): #<< สืบทอด
    def say_hello(self):
        print('Hello world')
class Truck(Vehicle):
    pass
class Van(Vehicle):
    pass

car1 = Car()
car1.turn_on_air()
car1.say_hello()

truck1 = Truck()
truck1.turn_on_air()

van1 = Van()
van1.turn_off_air()

```
